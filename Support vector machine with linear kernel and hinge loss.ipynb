```python
#Import libraries
import random
import math
```

```python
#Read data
# Open the file named 'ionosphere.data' in read mode
f = open("ionosphere.data")
# Initialize an empty list to store the data
x = []
# Initialize a counter variable to keep track of the number of lines read
i = 0
# Read the first line from the file
l = f.readline()
# Continue reading the file until the end is reached (i.e, l becomes an empty string)
while(l != ''):
    # Split the current line into a list of strings separated by space
    a = l.split()
    # Iterate over each element in the list 'a', convert it to float and append to l2
    l2 = []
    for j in range(0,len(a),1):
        l2.append(float(a[j]))
    x.append(l2)
    # Read the next line from the file
    l = f.readline()
    # Append the value '1' to the end of the current list in 'x' to include the bias term
    x[i].append(1)
    i +=1

#Calculate number of lines read from file and the number of columns 
rows = len(x)
cols = len(x[0])
f.close() 
```

```python
#Read labels
# Open the labels file in read mode
f = open("ionosphere.trainlabels.0")
# Initialize an empty dictionary to store the labels
y = {}
# Initialize an empty list to store the count of labels
n = []
# Append 0 to the list (for counting label 0)
n.append(0)
# Append another 0 to the list (for counting label 1)
n.append(0)
l = f.readline()
# Continue reading the file until the end is reached
while(l != ''):
    a = l.split()
    # Store the label (converted to integer) in the dictionary with the row index as the key
    y[int(a[1])] = int(a[0])
    # If the label is 0, convert it to -1
    if(y[int(a[1])] == 0):
        y[int(a[1])] = -1
    l = f.readline()
```

```python
# Initialize weights
import random
# Initialize an empty list to store the weights
w = []
# Initialize the weights to zero
for j in range(0,cols,1):
    w.append(0)  
# Assign a random weight each element in 'w'
for j in range(0,cols,1):
    w[j] = 0.02*random.uniform(0,1)-0.01
```

```python
#Dotproduct function
def dot(a,b):
    """
    Compute the dot product of two lists or vectors.
    
    Parameters:
    a (list or vector): The first vector.
    b (list or vector): The second vector.
    
    Returns:
    dp (float): The dot product of vectors 'a' and 'b'.
    """
    dp = 0
    for i,j in zip(a,b):
        dp += i*j
    return dp
```

```python
eta = 0.001 # Learning rate for gradient descent
tol = 0.001 # Stopping criterion 
C = 0.01 # Regularization parameter

# Calculate the L2 norm (Euclidean norm) of the weight vector (excluding the last weight)
normw = 0
for j in range(cols-1):
    normw += w[j]**2 # Sum the squares of the weights

# Calculate the objective function with hinge loss and add the regularization term
obj = 0
for i in range(0,rows,1):
    if(y.get(i) != None):
        for j in range(cols-1):
            obj += max(0, 1 - y[i]*dot(w,x[i])) + 0.5*C*(normw)
```

```python
# Initialize previous objective value to infinity
prev = float("inf")

# Loop until the change in objective function is less than stopping criterion
while(abs(prev - obj) > tol):
    prev = obj
    grad = []
    
    # Initialize gradients for all weights to zero
    for j in range(0,cols,1):
        grad.append(0)
        
    # Compute the gradient of the objective function    
    for i in range(0,rows,1):
        if(y.get(i) != None):
            # Update gradient for the weights based on the hinge loss
            for j in range(0,cols,1):
                if(y[i]*dot(w,x[i])) < 1:
                    grad[j] += -1*x[i][j]*y[i] + C*w[j]
                else:
                    grad[j] += C*w[j]
    
    # Adjust weights in the direction of the negative gradient
    for j in range(0,cols,1):
        w[j] = w[j] - eta * grad[j]
        
    # Recalculate the L2 norm of the weight vector    
    normw = 0
    for j in range(cols-1):
        normw += w[j]**2   
        
    # Recalculate the objective function value     
    obj = 0
    for i in range(0,rows,1):
        if(y.get(i) != None):
            for j in range(cols-1):
                obj += max(0, 1 - y[i]*dot(w,x[i])) + 0.5*C*(normw)
```

```python
# Classify unlabeled data using the trained model
for i in range(rows):
    if(y.get(i) == None):
        # Compute the dot product of the weight vector and the feature vector
        dp = dot(w,x[i])
        # If the dot product is positive, classify as class 1 else 0
        if(dp > 0):
            print(1,i)
        else:
            print(0,i)
```
